```python
match_inn1_6=F_1_6_ining_df.groupby(by=['matchId','battingTeam'], as_index=False).sum()
match_inn1_10=F_1_10_ining_df.groupby(by=['matchId','battingTeam'], as_index=False).sum()
match_inn1_15=F_1_15_ining_df.groupby(by=['matchId','battingTeam'], as_index=False).sum()
match_inn1_20=F_1_20_ining_df.groupby(by=['matchId','battingTeam'], as_index=False).sum()
```

```python
match_inn2_6=S_2_6_ining_df.groupby(by=['matchId','battingTeam'], as_index=False).sum()
match_inn2_10=S_2_10_ining_df.groupby(by=['matchId','battingTeam'], as_index=False).sum()
match_inn2_15=S_2_15_ining_df.groupby(by=['matchId','battingTeam'], as_index=False).sum()
match_inn2_20=S_2_20_ining_df.groupby(by=['matchId','battingTeam'], as_index=False).sum()
```

```python
not_found_1_6_inn=[]
for id1 in match_inn1_6.matchId.tolist():
    lst=int(id1[-1])+1
    id2=id1[:-1]+(str(lst))
    if id2 not in match_inn2_6.matchId.tolist():
      
        not_found_1_6_inn.append(id1)


not_found_1_10_inn=[]
for id1 in match_inn1_10.matchId.tolist():
    lst=int(id1[-1])+1
    id2=id1[:-1]+(str(lst))
    if id2 not in match_inn2_10.matchId.tolist():
        
        not_found_1_10_inn.append(id1)


not_found_1_15_inn=[]
for id1 in match_inn1_15.matchId.tolist():
    lst=int(id1[-1])+1
    id2=id1[:-1]+(str(lst))
    if id2 not in match_inn2_15.matchId.tolist():
        
        not_found_1_15_inn.append(id1)


not_found_1_20_inn=[]
for id1 in match_inn1_20.matchId.tolist():
    lst=int(id1[-1])+1
    id2=id1[:-1]+(str(lst))
    if id2 not in match_inn2_20.matchId.tolist():
        
        not_found_1_20_inn.append(id1)
```

```python
groupedBy_6=mergedInn1_6_Inn2_6.groupby(by=['battingTeam_x','key_0'], as_index=False).sum()

groupedBy_10=mergedInn1_10_Inn2_10.groupby(by=['battingTeam_x','key_0'], as_index=False).sum()

groupedBy_15=mergedInn1_15_Inn2_15.groupby(by=['battingTeam_x','key_0'], as_index=False).sum()

groupedBy_20=mergedInn1_20_Inn2_20.groupby(by=['battingTeam_x','key_0'], as_index=False).sum()
```

```python
wins_matches_6=groupedBy_6[groupedBy_6.runs_x > groupedBy_6.runs_y]
loss_matches_6=groupedBy_6[groupedBy_6.runs_x < groupedBy_6.runs_y]
draw_matches_6=groupedBy_6[groupedBy_6.runs_x == groupedBy_6.runs_y]

wins_matches_10=groupedBy_10[groupedBy_10.runs_x > groupedBy_10.runs_y]
loss_matches_10=groupedBy_10[groupedBy_10.runs_x < groupedBy_10.runs_y]
draw_matches_10=groupedBy_10[groupedBy_10.runs_x == groupedBy_10.runs_y]

wins_matches_15=groupedBy_15[groupedBy_15.runs_x > groupedBy_15.runs_y]
loss_matches_15=groupedBy_15[groupedBy_15.runs_x < groupedBy_15.runs_y]
draw_matches_15=groupedBy_15[groupedBy_15.runs_x == groupedBy_15.runs_y]

wins_matches_20=groupedBy_20[groupedBy_20.runs_x > groupedBy_20.runs_y]
loss_matches_20=groupedBy_20[groupedBy_20.runs_x < groupedBy_20.runs_y]
draw_matches_20=groupedBy_20[groupedBy_20.runs_x == groupedBy_20.runs_y]
```

```python
wins_matches_6['Winning Batting_Team'] =  1
wins_matches_10['Winning Batting_Team'] =  1
wins_matches_15['Winning Batting_Team'] = 1
wins_matches_20['Winning Batting_Team'] =  1


loss_matches_6['Winning Batting_Team'] =  0
loss_matches_10['Winning Batting_Team'] =  0
loss_matches_15['Winning Batting_Team'] =  0
loss_matches_20['Winning Batting_Team'] =  0

draw_matches_6['Winning Batting_Team'] = 1
draw_matches_10['Winning Batting_Team'] = 1
draw_matches_15['Winning Batting_Team'] = 1
draw_matches_20['Winning Batting_Team'] = 1
```

```python
overall_df_6 = shuffle(pd.merge(pd.merge(wins_matches_6, loss_matches_6,how='outer'),draw_matches_6,how='outer'))
overall_df_10 = shuffle(pd.merge(pd.merge(wins_matches_10, loss_matches_10,how='outer'),draw_matches_10,how='outer'))
overall_df_15 = shuffle(pd.merge(pd.merge(wins_matches_15, loss_matches_15,how='outer'),draw_matches_15,how='outer'))
overall_df_20 = shuffle(pd.merge(pd.merge(wins_matches_20, loss_matches_20,how='outer'),draw_matches_20,how='outer'))

overall_df_6.reset_index(drop=True,inplace=True)
overall_df_10.reset_index(drop=True,inplace=True)
overall_df_15.reset_index(drop=True,inplace=True)
overall_df_20.reset_index(drop=True,inplace=True)
```
